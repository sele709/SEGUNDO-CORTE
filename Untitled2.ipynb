# Librería Pandas en Python

```python
import pandas as pd
import numpy as np

# Crear DataFrame
datos = {
    "Producto": ["Arroz", "Leche", "Pan"],
    "Precio": [2000, 3000, 1500],
    "Cantidad": [2, 1, 3]
}

df = pd.DataFrame(datos)
print(df)

# Operación entre columnas
df["Total"] = df["Precio"] * df["Cantidad"]
print(df)

# Crear CSV y leerlo
df.to_csv("archivo.csv", index=False)

df_csv = pd.read_csv("archivo.csv")
print(df_csv.head())

# Datos sintéticos
datos = {
    "Edad": np.random.randint(18, 60, 10),
    "Salario": np.random.randint(1000, 5000, 10)
}

df_sintetico = pd.DataFrame(datos)
print(df_sintetico)
```
