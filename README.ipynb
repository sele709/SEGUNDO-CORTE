# **Taller Pandas- Analisis de Ventas**

**Descripción**

Este taller consiste en la manioulacion y analisis de datos de un conjunto de productos con sus precios, cantidades, costos y categorias, utilizando la libreria Pandas de Python.

**Objetivos**


1.  Aplicar medidas estadisticas descriptivas: media,mediana,moda. desviacion estándar, valores maximos y minimos.
2.  Filtrar y ordenar datos según categorías y criterios especificos.
3.   Interpretar resultados y comprender su importancia en la toma de deciciones catégoricas.
4.  Desarrollar habilidades en Python para el análisis de datos de manera organizada y reproducible






**Metodologia**

1. **Carga y revision del dataset:**
se trabajo con un archivo de ventas en excel que contiene columnas como Producto, Precio, Cantidad, Categoria y Total_Venta
2. E**stadisticas Descriptivas:** Se calcularon medidas centrales y de dispersión para comprender distribucción de los precios y la cantidad de productos vendidos.
Estas medidas permiten:


*   **Media** Promedio de los precios o cantidades, representando al valor típico.
*   **Mediana**: Valor central que divide los datos en dos partes iguales, evitando el sesgo por valores extremos.
* **Moda**: El valor más frecuente, útil para identificar tendencias de consumo.
* **Desviación estándar**: Indica cuánto se dispersan los datos respeto a la media.
* **Valores maximos y minimos:** Muestran los límites del rango de ventas.
3. **Filtrado y ordenamiento:** Se selecionaron los productos de la categoria "Tecnologia" y se ordenaron segun el Total_Venta de mayor a menor. Ademas se trabajo especificamente con los primero 5 y los ultimos 10 productos para observar patrones de alto y bajo desempeño de ventas.
4. **Interpretación de resultados:** Cada resultado fue analizado para general conclusiones sobre el comportamiento de los productos y su rentabilidad de priorizar productos estrategicos de decisiones comerciales.
